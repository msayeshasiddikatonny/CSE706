```python
import time
import os
from memory_profiler import memory_usage

def single_loop():
    """CPU Bound: 10 Million iterations"""
    count = 0
    for i in range(10**7):
        count += i
    return count

def nested_loop():
    """CPU Bound: 2500 x 2500 iterations"""
    count = 0
    for i in range(2500):
        for j in range(2500):
            count += (i * j)
    return count

def memory_heavy_task():
    """Memory Bound: Allocating a large list of floats"""
    large_list = [float(i) for i in range(5 * 10**6)]
    del large_list # Clean up

def io_bound_task():
    """I/O Bound: Simulating a 1-second network/disk delay"""
    time.sleep(1.0)

def run_benchmarks():
    tasks = {
        "SingleLoop": single_loop,
        "NestedLoop": nested_loop,
        "MemoryHeavy": memory_heavy_task,
        "IOBoundTask": io_bound_task
    }

    print(f"{'Task':<20} | {'Time (ms)':<15} | {'Memory (MB)':<15}")
    print("-" * 55)

    for name, func in tasks.items():
        # Measure Memory
        mem_before = memory_usage(-1)[0]

        # Measure Time
        start_time = time.perf_counter()
        func()
        end_time = time.perf_counter()

        mem_after = memory_usage(-1)[0]
        duration_ms = (end_time - start_time) * 1000
        mem_diff = max(0, mem_after - mem_before)

        print(f"{name:<20} | {duration_ms:<15.4f} | {mem_diff:<15.4f}")

if __name__ == "__main__":
    run_benchmarks()
```
